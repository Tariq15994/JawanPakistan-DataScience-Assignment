```python
attack_mean = data["Attack"].mean()
```

```python
def set_attack(val):
    if val< attack_mean:
        return "Attack High"
    elif val==attack_mean:
        return "Attack Neutral"
        
    else:
        return "Attack Low"
```

```python
data["Attack_high_low"] = data["Attack"].apply(set_attack)
```

```python
data.head()
```

```python
speed_mean = data["Speed"].mean()
def set_speed(val):
    if val< speed_mean:
        return "Speed low"
    elif val==speed_mean:
        return "Speed Neutral"
        
    else:
        return "Speed High"
```

```python
speed_mean
```

```python
data["Speed_High_Low"] = data["Speed"].apply(set_speed)
```

```python
data['Speed_High_Low']
```

```python
hp_mean = data["HP"].mean()
def set_hp(val):
    if val< hp_mean:
        return "HP low"
    elif val==hp_mean:
        return "HP Neutral"
        
    else:
        return "HP High"
```

```python
data["HP_High_Low"] = data["HP"].apply(set_hp)
```
